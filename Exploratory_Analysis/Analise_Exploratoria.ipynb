# Análise Exploratória de Dados

## Contexto

#### **Variáveis**:

**customerID:** ID de cliente<br/>
**gender:** Género (Masculino - male | Feminino - female)<br/>
**SeniorCitizen:** Se o cliente é um cidadão sénior (1, 0)<br/>
**Partner:** Se o cliente tem um(a) parceiro(a) (Yes, No)<br/>
**Dependents:** Se o cliente tem dependentes (Yes, No)<br/>
**tenure:** Número de meses desde que o cliente "entrou" na empresa<br/>
**PhoneService:** Se o cliente tem serviço de telefone (Yes, No)<br/>
**MultipleLines:** Se o cliente tem linhas múltiplas (Yes, No, No phone service)<br/>
**InternetService:** Tipo de ligação à internet (DSL, Fiber optic, No)<br/>
**OnlineSecurity:** Se o cliente contratou o serviço de segurança na internet (Yes, No, No internet service)<br/>
**OnlineBackup:** Se o cliente contratou o serviço de cópia de segurança *onilne* (Yes, No, No internet service)<br/>
**DeviceProtection:** Se o cliente contratou o serviço de protecção de dispositivos (Yes, No, No internet service)<br/>
**TechSupport:** Se o cliente contratou o serviço de suporte técnico (Yes, No, No internet service)<br/>
**StreamingTV:** Se o cliente contratou o serviço de *streaming* de TV (Yes, No, No internet service)<br/>
**StreamingMovies:** Se o cliente contratou o serviço de *streaming* de filmes (Yes, No, No internet service)<br/>
**Contract:** Duração do contrato (Month-to-month, One year, Two year)<br/>
**PaperlessBilling:** Se o cliente contratou o serviço de factura electrónica (Yes, No)<br/>
**PaymentMethod:** Método de pagamento (Electronic check, Mailed check, Bank transfer (automatic), Credit card (automatic))<br/>
**MonthlyCharges:** Cobrança mensal ao cliente<br/>
**TotalCharges:** Cobrança total ao cliente<br/>
**Churn:** Se o cliente deixou a empresa (Yes or No)

Apenas uma variável tem valores em falta: *TotalCharges* - faltam 11 valores, como vimos anteriormente.

---

```python
cust_data['TotalCharges_no_miss'] = cust_data['TotalCharges'].fillna(cust_data['tenure']*cust_data['MonthlyCharges'])
```

Uma vez que sabemos há quantos meses o cliente "está" na empresa e não há nenhum valor em falta na cobrança mensal, talvez multiplicar a cobrança mensal (*MonthlyCharges*) pelo número de meses (tenure) nos dê um valor plausível.

Em vez de substituir directamente na variável original, podemos criar uma variável nova, uma vez que a original pode, ainda assim, ser utilizada em alguns modelos (como *random forest*, por exemplo).

```python
print(cust_data.agg({'TotalCharges': ['mean', 'median']}))
cust_data.TotalCharges.plot(kind='hist')
plt.show()
```

```python
sns.pairplot(cust_data, hue = 'Churn')
```

As variáveis *tenure* e *MonthlyCharges* para os 'churners' apresentam distribuições diferentes dos clientes que se mantêm. Estas duas variáveis poderão ser bons preditores do abandono.

---

```python
categoricas = cust_data.select_dtypes('object').columns[2:-1] # var auxiliar, para loop
for cat in categoricas:
    print(cust_data.groupby(cat)['Churn'].value_counts(normalize=True))
```

Solução com mais informação e de forma visual

Alternativa:

```python
for cat in categoricas:
    tmp = cust_data.groupby(cat)['Churn'].value_counts(normalize=True)#.unstack()
    print(f'\033[1m{cat}\033[0m') # prints variable name in bold ('\033[1m' inicia o bold e '\033[0m' termina)
    print(tmp.unstack())
    sns.barplot(data=tmp.rename('%').reset_index(), x='%', y=cat, hue='Churn')#tmp.plot.barh()
    plt.show()
    print('\n') # add space
```

Variáveis onde se notam diferenças maiores na % de abandono vs não-abandono:
- 'SeniorCitizen'
- 'InternetService'
- 'OnlineSecurity'
- 'OnlineBackup'
- 'DeviceProtection'
- 'TechSupport'
- 'Contract'
- 'PaymentMethod'
- 'PaperlessBilling'

Variáveis onde existe diferença mas não é tão acentuada:
- 'Partner'
- 'Dependents'

Variáveis sem diferenças muito acentuadas:
- 'Gender'
- 'PhoneService'
- 'MultipleLines'
- 'StreamingTV'
- 'StreamingMovies'

**Nota:** No caso das variáveis 'StreamingTV' e 'StreamingMovies', notam-se algumas diferenças, mas apenas no grupo 'No internet service' (em ambos os casos). Esse facto sugere que a 'verdadeira' influência é da variável 'InternetService', pelo que optamos por não considerar estas no grupo das influentes.
